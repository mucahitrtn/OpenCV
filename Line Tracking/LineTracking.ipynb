```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("C:\\Users\\canis\\hsvhsl.png")
```

```python
cv2.imshow("Orginal image", img)
```

```python
def convert(img, color):
    if color=="hsv":
        img=cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    elif color=="hls":
        img=cv2.cvtColor(img, cv2.COLOR_BGR2HLS)
    elif color=="grayscale":
        img=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    return img
```

```python
hsl= convert(img, "hls")
hsv= convert(img, "hsv")
```

```python
gray= convert(hsl, "grayscale")
plt.imshow(gray,"gray")
```

```python
size= (4,2)
blur= cv2.blur(gray,size)
plt.imshow(blur, "gray")
```

```python
height, widht= blur.shape
can= cv2.Canny(gray,75,150)
canny= cv2.Canny(blur, 100,200)
plt.imshow(canny,"gray")
edges= cv2.Canny(hsv, 70,150)
```

```python
slicedimg = edges[170:370, 1:400]
```

```python

plt.imshow(slicedimg, "gray")
```

```python
plt.imshow(slicedimg,"gray")
```

```python
lines= cv2.HoughLinesP(slicedimg, 1, np.pi/180, 100 , maxLineGap=50)
```

```python
for line in lines:
    x1,x2,y1,y2= line[0]
    cv2.line(slicedimg,(x1,y1),(x2,y2), (255,0,0),5)
```
