```python
import cv2
import numpy as np 
```

```python
img = cv2.imread("C:\\Users\\canis\\hsvhsl.png")
```

```python
cv2.imshow("Orginal image", img)
```

```python
def convert(img, color):
    if color=="hsv":
        img=cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    elif color=="hls":
        img=cv2.cvtColor(img, cv2.COLOR_BGR2HLS)
    elif color=="grayscale":
        img=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    return img
```

```python
hsl= convert(img, "hls")
cv2.imshow("converted to hls", hsl)
hsv= convert(img, "hsv")
cv2.imshow("converted to hsv", hsv)
gray= convert(img, "grayscale")
cv2.imshow("converted to grayscale", gray)
cv2.waitKey(0)
```
